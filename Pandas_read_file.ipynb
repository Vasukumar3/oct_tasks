```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

eq_data = pd.read_csv("earthquake_data_tsunami.csv")
# print(eq_data.info())
# print(eq_data.describe())
print(eq_data.head())
print(eq_data.tail())

eq_data.isnull().sum()

tsunami_yrs_mnth = eq_data[eq_data.tsunami > 0]
print(tsunami_yrs_mnth[['Year','Month','tsunami']])

#group by and count of tusnami in each year
tsunami_yrs = eq_data.groupby(['Year'])['tsunami'].count()
print(tsunami_yrs)

x_axis = tsunami_yrs.index
y_axis = tsunami_yrs.values

#tsunami_yrs.plot(x='Year',y='tsunami',marker='o')
plt.subplot(1,2,1)
plt.plot(x_axis,y_axis,'o')
plt.title("Tsunami occurrences over years")
plt.xlabel("Year of Earthquake occurrence")
plt.ylabel("Tsunami occurrence")

plt.subplot(1,2,2)
# tsunami_yrs.plot(x='Year',y='tsunami',marker='o')
plt.bar(x_axis,y_axis, color ='r')
plt.title("Tsunami occurrences over years")
plt.xlabel("Year of Earthquake occurrence")
plt.ylabel("Tsunami occurrence")

plt.grid()
plt.show()
```
